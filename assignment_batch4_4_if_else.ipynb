### 44. Check if a list contains duplicates.

```python
ls=[1,2,4,12,1,4,1,42]
seen=set()
duplicate=False
for i in ls:
    if i in seen:
        duplicate=True
        break
    seen.add(i)
if duplicate:
    print("List contains duplicate")
else:
    print("List don't have any duplicates")
```
