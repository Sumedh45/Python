# Basic List Operations 

1. Create a list of 5 numbers and add a new number to the end.


```python
ls=[1,2,3,4,5]
print(ls)
ls.append(6)
ls
```

2. Make a list of fruits and insert a fruit at the second position.


```python
fruits=['apple','banana','pineapple',"watermelon"]
print(fruits)
fruits.insert(1,'mango')
fruits
```

3. Create a list of names and remove one name from the list.


```python
name=['sumedh','ajinkya','pavan','gaurav']
print(name)
name.remove('gaurav')
name
```

4. Create a list of 4 items and change the value at index 2.


```python
items=['abc','xyz','pqr','stu']
print(items)
items[2]='efg'
items
```

5. Make a list of numbers and print only the first three elements.


```python
fruits=['apple','banana','pineapple',"watermelon",'guava']
print(fruits)
fruits[:3]
```

6. Create a list of colors and find the index of a specific color.


```python
color=['red','blue','grey','black','white','orange']
color.index('blue')
```

7. Create a list of 10 numbers and print how many times “5” appears.


```python
n=[1,2,4,5,3,2,4,2,2,3,4,5,6,7,7,6,5,4,65,44,3,12,4,5,5,3,3]
n.count(5)
```

8. Make a list of 6 items and delete the item at index 3.


```python
color=['red','blue','grey','black','white','orange']
print(color)
color.pop(3)
color
```

9. Create a list of animals and add three more animals using extend().


```python
animal=['cat','tiger','dog']
animal1=['horse','cow','deer']
print(animal)
animal.extend(animal1)
animal
```

10. Create a list and remove the last item using pop().


```python
animal=['cat', 'tiger', 'dog', 'horse', 'cow', 'deer']
print(animal)
animal.pop()
animal
```

11. Create a list of prices and sort them in ascending order.


```python
prices=[123,3453,23426,232,235,121]
prices.sort()
prices
```

12. Make a list of countries and reverse the order of the list.


```python
countries = ['Canada', 'Brazil', 'Germany', 'Japan', 'India', 'South Africa']
print(countries)
countries.reverse()
countries
```

14. Create a list of numbers and print the largest and smallest number.


```python
prices=[123,3453,23426,232,235,121]
print(prices)
print('Max- ',max(prices))
print('Min- ',min(prices))
```
