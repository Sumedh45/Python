1. Create a dictionary with 3 key-value pairs (name, age, city).

2. Create a dictionary and print the value of a specific key.


```python
di={i:i*i for i in range(1,6)}
print(di)
di[2]
```

3. Make a dictionary and add a new key-value pair to it.


```python
di={i:i*i for i in range(1,6)}
print(di)
di[6]=36
di
```

4. Create a dictionary and change the value of an existing key.


```python
di={i:i*i for i in range(1,6)}
print(di)
di[5]=22
di
```

5. Create a dictionary and remove one key using pop().


```python
print(di)
di.pop(5)
di
```

6. Create a dictionary and remove the last inserted item using popitem().


```python
di={i:i*i for i in range(1,6)}
print(di)
di.popitem()
di
```

7. Create a dictionary and print all the keys.


```python
di={i:i*i for i in range(1,6)}
print(di)
for key in di.keys():
    print(key)
```

8. Create a dictionary and print all the values.


```python
di={i:i*i for i in range(1,6)}
print(di)
for val in di.values():
    print(val)
```

9. Create a dictionary and print all the key-value items.


```python
di={i:i*i for i in range(1,6)}
print(di)
for key,val in di.items():
    print(key,val)
```

10. Make a dictionary and check the number of items using len().
